# Differentiation game: pilot experiment results

## Structure of the notebook 
##### 1. Load the dataset
##### 2. Display descriptive statistics
##### 3. Display plots

## 1. Load the dataset 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

player = pd.read_csv('data/player.csv')
print(player)
```

## 2. Display descriptive statistics

```python
# Count of participants in each 'high_reward' group
high_reward_counts = player['high_reward'].value_counts()
print("Counts of Participants for Each High Reward Group:")
print(high_reward_counts)

# Distribution of 'more_different' within each 'high_reward' group
more_different_distribution = player.groupby('high_reward')['more_different'].value_counts(normalize=True)
print("\nDistribution of 'More Different' Responses Within Each 'High Reward' Group:")
print(more_different_distribution.unstack()*100)  # Converting proportions to percentages for easier interpretation
```

## 3. Display plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Plot setup
plt.figure(figsize=(16, 8))  # Increase figure size to accommodate legend outside

# Plot 1: Bar Plot for 'high_reward' Counts
sns.countplot(x='high_reward', data=player)
plt.title('Number of Participants per High Payoff Group\n', fontsize=18)
plt.xlabel('\nHigher Payoff (Condition)\n', fontsize=18)
plt.ylabel('Count', fontsize=18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show() 

# Prepare data for Plot 2
more_different_distribution = player.groupby('high_reward')['more_different'].value_counts(normalize=True).unstack().fillna(0)

# Plot 2: Stacked Bar Plot for 'more_different' Distribution
ax = more_different_distribution.plot(kind='bar', stacked=True, color=['skyblue', 'lightgreen'], figsize=(16, 8))
plt.title('For a Given Higher Payoff Group, Distribution of \'More Different\' Responses\n', fontsize=20)
plt.xlabel('Higher Payoff Group (Condition)', fontsize=18)
plt.ylabel('Percentage\n', fontsize=18)
plt.xticks(fontsize=18, rotation=0)  # Adjust rotation if necessary
plt.yticks(fontsize=18)
# Move legend to the right of the plot
plt.legend(title='"Which array is more different"?', title_fontsize='12', loc='upper left', fontsize=10, bbox_to_anchor=(1.05, 1))

# Adjust overall layout and show plots
plt.tight_layout()

# Save the figure
plt.savefig('results.png', dpi=300, bbox_inches='tight')
plt.show()
```
